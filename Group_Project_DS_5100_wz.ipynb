```python
# group project DS5100 import database,
#site to arrange data by dates:https://www.interviewqs.com/ddi-code-snippets/select-pandas-dataframe-rows-between-two-dates
import numpy as np
import pandas as pd
import seaborn as sns

NYTCovid= pd.read_csv("https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv")

NYTCovid['date'] = pd.to_datetime(NYTCovid['date']) #make sure is in date format.
```

```python

start_date = '03-01-2020'
end_date = '09-01-2020'
mask = (NYTCovid['date'] > start_date) & (NYTCovid['date'] <= end_date)
#assign mask to df to return the rows with _date between our specified start/end dates

NYTCovid_peakdates = NYTCovid.loc[mask] #specified the date range to be within 1st 6 months of pandemic onset.
NYTCovid_peakdates
```
